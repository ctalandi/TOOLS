```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import matplotlib as mpl
from matplotlib import ticker

%matplotlib inline
```

```python
# Grid depth and resolution
z=xr.open_dataset('ICE_canal_deptht.nc')
Dx = 2e3
z
```

```python
def zdf(av_surf,depth):
    # initialization of the profile
    zratio = 0.25           # surface/bottom ratio
    zh =  500.              # depth    of the inflection point [m]
    zw =  1. / 200.         # width^-1     -        -      -   [1/m]
    #                       # associated coefficient           [-]
    zc = ( 1. - zratio ) / ( 1. + np.tanh( zh * zw) )

    pah1 = av_surf * (  zratio + zc * ( 1. + np.tanh( - ( depth - zh ) * zw) )  )
    return pah1
```

```python
# Laplacian viscosity 
rn_Uv= 0.3  ;  
ahm = 1/2 *rn_Uv*Dx
print("ahm=",ahm)
# bi-Laplacian diffusivity
rn_Uv= 0.03  ; Dx = 2e3 
aht = 1/12 *rn_Uv*Dx**3
print("aht=",aht*1e-6)
```

```python
plt.figure(figsize=(20,15))

plt.subplot(211)
plt.plot(np.array(z['deptht']),zdf(ahm,np.array(z['deptht'])) )
plt.grid(True)
plt.xlabel('Depth')
plt.ylabel('Ahm (m2/s)')
plt.title('Laplacian viscosity')

plt.subplot(212)
plt.plot(np.array(z['deptht']),zdf(aht,np.array(z['deptht'])) )
plt.grid(True)
plt.xlabel('Depth')
plt.ylabel('Aht (m4/s)')
plt.title('bi-Laplacian diffusivity')
```
