```python
def htau(gphit):
    
    hsin= 45.* np.abs( np.sin( np.pi/180. * gphit ) )
    htau1=np.zeros((2,gphit.size))
    htau1[0,:]=30.
    htau1[1,:]=hsin
    htau2=np.min(htau1,axis=0)

    htau05=np.zeros((2,gphit.size))
    htau05[0,:]=0.5
    htau05[1,:]=htau2
    htau3=np.max(htau05,axis=0)
    
    return htau3
```
