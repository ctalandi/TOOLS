```python
#!/usr/bin/env python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
from matplotlib import rcParams
from mpl_toolkits.basemap import Basemap
import matplotlib as mpl
import xarray as xr 
import shapiro
```

```python
# Print arguments to pass to the shapiro function
print(shapiro.lisshapiro1d.__doc__)
```

```python
os.environ['LD_LIBRARY_PATH'] = os.environ['LD_LIBRARY_PATH'] +':/opt/linux/intel/12.1/lib/intel64'
print(os.environ['LD_LIBRARY_PATH'])
```

```python
# Read the input data field 

ds=xr.open_dataset('/data0/project/drakkar/CONFIGS/CREG12.L75/CREG12.L75-REF08-S/5d/2000/CREG12.L75-REF08_y2000m08d18.5d_gridT.nc')
dsgrd=xr.open_dataset('/data0/project/drakkar/CONFIGS/CREG12.L75/GRID/CREG12.L75-REF08_mask.nc')
dscoo=xr.open_dataset('/data0/project/drakkar/CONFIGS/CREG12.L75/GRID/coordinates_CREG12.L75.nc')
ds
dsgrd
temp=np.array(ds['ssh'][0,:,:].squeeze())
```

```python
dsgrd
```

```python
plt.figure(figsize=(20,15))
dscoo['rp_aniso']=('y','x'),xr.where(dsgrd['tmask'][0,0,:,:] == 0., np.nan,dscoo['e1t']/dscoo['e2t']*dsgrd['tmask'][0,0,:,:])
dscoo['rp_aniso'].plot(vmin=0.5,vmax=1.5,levels=40,cmap='Spectral_r')
plt.contour(dscoo['e1t']/dscoo['e2t']*dsgrd['tmask'][0,0,:,:],contours=[1.],colors='magenta')
```

```python
# The transpose is required to pass from FROTRAN to PYTHON dimension order 
ssh=xr.where(np.isnan(np.array(ds['ssh'][0,:,:])), 0., np.array(ds['ssh'][0,:,:].squeeze()))

#ssh=np.array(ds['ssh'][0,:,:].squeeze().T)
msk=np.array(dsgrd['tmask'][0,0,:,:].squeeze().T)
print(temp.shape)
print(ssh.shape)
```

```python
plt.imshow(ssh)
```

```python
plt.imshow(msk)
```

```python
# Apply the Shapiro filter 
it_shap=50
out=shapiro.lisshapiro1d(ssh.T,msk,it_shap)
print(out.shape)
```

```python
# Save the new smoothed data 

ds_out=xr.Dataset()
ds_out['ssh_smo']=(('y','x'),out.T)
ds_out['ssh_diff']=(('y','x'),out.T-ssh)
ds_out
```

```python
# Plot the new field and show the differences with the nitial one.
plt.figure(figsize=(20,15))
ds_out['ssh_smo'].plot()
```

```python
plt.figure(figsize=(20,15))
ds_out['ssh_diff'].plot()
```

```python
# Save the new smoothed data 
ds_out.to_netcdf('./ssh_smooth_Shapiro'+str(it_shap)+'.nc')
```
